```python
# Importação de Biblioteca.
import pandas as pd
```

```python
# Carregar o CSV Prata.
df = pd.read_csv("../2-prata/dados_prata.csv")
```

```python
# Conversão do GDP_média de string "%" para float.
df["Média do Crescimento"] = df["Média do Crescimento"].str.replace("%", "", regex=False).astype(float)
```

```python
# Criação de um ranking de crescimento proporcional (%).
df["Ranking de crescimento"] = df["Média do Crescimento"].rank(ascending=False, method="dense").astype(int)
```

```python
# Classificação do crescimento dos países.
def classificar_crescimento(valor):
    if valor >= 10:
        return "Alto"
    elif valor >= 5:
        return "Médio"
    else:
        return "Baixo"
    
    
df["Faixa de crescimento"] = df["Média do Crescimento"].apply(classificar_crescimento)
```

```python
# Converter a coluna 2020 para float (ela está formatada com $ e ,).
df["2020_float"] = df["2020"].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Converter a coluna 2025 para float.
df["2025_float"] = df["2025"].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Calcula participação em 2020 (GDP do país em relação ao mundo).
total_2020 = df["2020_float"].sum()
df["Participação 2020"] = (df["2020_float"] / total_2020) * 100
```

```python
# Calcular participação em 2025.
total_2025 = df["2025_float"].sum()
df["Participação 2025"] = (df["2025_float"] / total_2025) * 100
```

```python
# Formatando participação 2020 em %.
df["Participação 2020"] = df["Participação 2020"].apply(lambda x: f"{x:.2f}%")
```

```python
# Formatando participação 2025 em %.
df["Participação 2025"] = df["Participação 2025"].apply(lambda x: f"{x:.2f}%")
```

```python
# Adição de uma coluna de tendência de participação.
def tendencia(p2020, p2025):
    p2020 = float(p2020.replace('%',''))
    p2025 = float(p2025.replace('%',''))
    if p2025 > p2020:
        return "↑ Cresceu"
    elif p2025 < p2020:
        return "↓ Caiu"
    else:
        return "→ Estável"

df["Tendência de Participação"] = df.apply(lambda row: tendencia(row["Participação 2020"], row["Participação 2025"]), axis=1)
```

```python
# Valor de crescimento do PIB bruto entre 2020 e 2025 (valor total).
df["Crescimento Bruto"] = df["2025_float"] - df["2020_float"]
df["Crescimento Bruto"] = df["Crescimento Bruto"].apply(lambda x: f"${x:,.0f}")
```

```python
# Criação de um Ranking de crescimento bruto.
df["Ranking de Crescimento Bruto"] = df["Crescimento Bruto"].rank(ascending=False, method="dense").astype(int)
```

```python
# Remoção de colunas auxiliares antes de salvar.
df = df.drop(columns=["2020_float", "2025_float"])
```

```python
# Salvar CSV Ouro.
df.to_csv("dados_ouro.csv", index=False)
```
